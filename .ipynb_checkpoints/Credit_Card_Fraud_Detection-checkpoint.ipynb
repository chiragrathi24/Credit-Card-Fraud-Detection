### Credit Card Fraud Detection

```python
#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
#Reading the data from the csv file
c_data = pd.read_csv('credit_card_data.zip')
```

```python
#Classifying the transactions
genuine = c_data[c_data['Class']==0]
fraud = c_data[c_data['Class']==1]
```

```python
#Counting the number of genuine and the fraud transactions
tansactions = ['genuine', 'fraud']
ccount = c_data['Class'].value_counts()
ccount
```

```python
#Plotting the pie chart of the transactions
labels = ['genuine', 'fraud']
plt.pie(ccount, labels = labels, autopct = '%.2f%%')
plt.title('Genuine vs Fraud transactions')
plt.show()
```

```python
#Prining the shape of the genuine and the fraud transactions
print('Genuine transactions shape = ', genuine.shape)
print('Fraud transactions shape = ', fraud.shape)
```

```python
#Plotting a Heat Map to find the correlations in between the different variables
plt.figure(figsize = (20,20))
sns.heatmap(c_data.corr(), annot = True, linewidths = 0.1)
```

```python
#Creating the input set
X = c_data.drop(['Class'], axis = 1)
X
```

```python
#Creating the output set
y = c_data['Class']
y
```

```python
#Splitting the data in the train and the test data sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```
